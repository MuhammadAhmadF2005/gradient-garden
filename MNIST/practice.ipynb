```python
import numpy as np
import matplotlib.pyplot as plt

# Fake data (y = 4 + 3x + noise)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.title("Training Data")
plt.show()
```

```python
theta_0 = 0.0
theta_1 = 0.0
learning_rate = 0.1
n_iterations = 1000
m = len(X)
```

```python
for iteration in range(n_iterations):
    y_pred = theta_0 + theta_1 * X
    
    error = y_pred - y
    
    # gradients
    d_theta_0 = (2/m) * np.sum(error)
    d_theta_1 = (2/m) * np.sum(error * X)

    if iteration % 100 == 0:
        print(theta_0, theta_1)
    
    # update
    theta_0 -= learning_rate * d_theta_0
    theta_1 -= learning_rate * d_theta_1
```

```python
print(theta_0, theta_1)

plt.scatter(X, y)
plt.plot(X, theta_0 + theta_1 * X, color='red')
plt.title("Fitted Line")
plt.show()
```
